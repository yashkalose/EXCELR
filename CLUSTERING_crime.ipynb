```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
data=pd.read_csv('crime_data.csv')
```

```python
data
```

```python
data.rename(columns={'Unnamed: 0':'States'},inplace=True)
```

```python
data
```

```python
data1=data.iloc[:,1:5]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df=scaler.fit_transform(data1)
```

HIERACHICAL

```python
dendrogram=sch.dendrogram(sch.linkage(df,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
```

```python
hc.fit(df)
```

```python
y_hc=hc.fit_predict(df)
```

```python
y_hc
```

```python
data['cluster_id']=y_hc
```

```python
df1=data.sort_values("cluster_id")
df1
```

```python
df1.groupby('cluster_id').agg(['mean']).reset_index()
```

KMEANS

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,random_state=0)
  kmeans.fit(df)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("ELBOW METHOD")
plt.xlabel("number of clusters")
plt.ylabel("wcss")
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
```

```python
predict=kmeans.fit_predict(df)
```

```python
predict
```

```python
data['kmeans_cluster_id']=predict
```

```python
data
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_score_average = silhouette_score(df, predict)
silhouette_score_average
```

```python
silhouette_score_average1 = silhouette_score(df, kmeans.predict(df))
silhouette_score_average1
```

DBSCAN

```python
dbscan = DBSCAN(eps=3, min_samples=3)
dbscan.fit_predict(df)
```

```python
dbscan.labels_
```

```python
data['DBSCAN_cluster']=dbscan.labels_
data
```
