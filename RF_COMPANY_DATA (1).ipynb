```python
cut_labels=['Low','Medium','High']
cut_bins=[-1,5.66,12,17]
data['sales']=pd.cut(data['Sales'],labels=cut_labels,bins=cut_bins)
```

```python
df=data.drop(columns=('Sales'),axis=1)
```

```python
df.head()
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df['ShelveLoc']=label_encoder.fit_transform(df['ShelveLoc'])
df['Urban']=label_encoder.fit_transform(df['Urban'])
df['US']=label_encoder.fit_transform(df['US'])
df['sales']=label_encoder.fit_transform(df['sales'])
```

```python
df.head()
```

```python
x = df.iloc[:,0:10]
y = df.iloc[:,10]
```

```python
num_trees = 100
max_features = 3
kfold = KFold(n_splits=10)
model = RandomForestClassifier(n_estimators=num_trees, max_features=max_features,random_state=0)
results = cross_val_score(model, x, y, cv=kfold)
print(results.mean())
```
