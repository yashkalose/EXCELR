```python
books=df1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
books
```

```python
books.fillna(0, inplace=True)
```

```python
books
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(books.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df1['User.ID'].unique()
user_sim_df.columns = df1['User.ID'].unique()
```

```python
user_sim_df
```

```python
#making diagonal elements zero
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:20, 0:20]
```
